```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
import pandas as pd

df = pd.read_csv (r'C:\Users\M Arsalan Siddiqui\OneDrive\Documents\Book4.csv')
df.head(11)
```

```python
X = df[['HEIGHT']].values
Y = df[['WEIGHT']].values
```

```python
def line(x , w=0 , b=0):
    return w*x +b
def mean_squared_error(Y , y_pred):
    s=(Y-y_pred)**2
    return np.sqrt(s.mean())
```

```python
plt.figure(figsize=(10,5))
ax1 = plt.subplot(121)
df.plot(kind='scatter',
       x='HEIGHT',
       y='WEIGHT',
       title = 'WEIGHT AND HEIGHT IN ADULTS',
       ax=ax1)
bs = np.array([-100,-50,0,50,100,150])
mses=[]
for b in bs:
    y_pred = line(X , w=2 , b=b)
    mse= mean_squared_error(Y, y_pred)
    mses.append(mse)
    plt.plot(X,Y)
    
ax2 = plt.subplot(122)
plt.plot(bs , mses, 'o-')
plt.title('cost as a function of b')
plt.xlabel('b')
```
